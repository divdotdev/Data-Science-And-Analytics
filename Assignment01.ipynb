```python
import numpy as np
m=int(input("Enter number of products:"))
cost=[]
for i in range(m):
    x=int(input("Enter price:"))
    cost.append(x)
cost=np.array(cost)
print("Cost Array =",cost)
print("Total Cost =",np.sum(cost))
print("Mean Cost =",np.mean(cost))
print("Lowest Cost =",np.min(cost))
print("Highest Cost =",np.max(cost))
print("Standard Deviation =",np.std(cost))
print("Sorted Cost =",np.sort(cost))
print("Distinct Cost =",np.unique(cost))
print("Cost Greater Than 100 =",cost[cost>100])
```
