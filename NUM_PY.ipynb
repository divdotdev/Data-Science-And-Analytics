```python
import numpy as np

a = np.random.randint(2, 15, 6)
b = np.random.randint(1, 19, 6)

print(a)
print(b)

c = np.where(b > 2, b, a)
print(c)

d = np.sort(a)
print(d)

e = np.sum(a)
print(e)

f = np.mean(a)
print(f)

g = np.std(a)
print(g)

h = np.min(a)
print(h)

i = np.max(a)
print(i)

j = np.unique(np.array([1, 2, 3, 2, 4, 1, 5, 3]))
print(j)

m1 = np.array([[1, 2],
               [3, 4]])

m2 = np.array([[5, 6],
               [7, 8]])

k = np.matmul(m1, m2)
print(m1)
print(m2)
print(k)
```
