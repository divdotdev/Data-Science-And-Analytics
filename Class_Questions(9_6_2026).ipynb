```python
#Question 1:

# Create a 5x5 matrix of random integers between (1-90)
# Find mean of values rows and columns wise
# Replace all even numbers with -1.
# Sort the array row and column wise
# Find row and column wise sum seperately
# Count the number of -ve's in each column
```

```python
import numpy as np
```

```python
# Create a 5x5 matrix of random integers between (1-90)v

matrix=np.random.randint(1,91,size=(5,5))
print(matrix)
```

```python
# Replace all even numbers with -1.

even_matrix=np.where(matrix%2==0,-1,matrix)
print(even_matrix)
```

```python
# Find mean of values rows and columns wise
rows=np.mean(matrix,1)
columns=np.mean(matrix,0)
print(rows)
print(columns)
```

```python
# Sort the array row and column wise
sort_rows=np.sort(matrix,1)
sort_columns=np.sort(matrix,0)
print(sort_rows)
print(sort_columns)
```

```python
# Find row and column wise sum seperately

rows_sum=np.sum(matrix,1)
columns_sum=np.sum(matrix,0)
print(rows_sum)
print(columns_sum)
```

```python
# Count the number of -ve's in each column

negative_cnt=np.sum(matrix<0)
print(negative_cnt)

count_col=np.sum(np.where(even_matrix<0,1,0),axis=0)
print(count_col)
```

```python
# Question 2:

# Inventory Control System

# Task:

# Create a 5x5 array (stock levels)
# Create anothe 5x5 array (demand)
# Compute:
# Remaining stock = stock - demand
# Replace:
# Negative values -> 0(Out of Stock)
# Find:
# Total Storage
# Products with zero stock
# Sort each row based on remaining stock
```

```python
import numpy as np
```

```python
# Create a 5x5 array (stock levels)
stock=np.random.randint(20,80,size=(5,5))
print(stock)
```

```python
# Create a 5x5 array (demand)
demand=np.random.randint(20,80,size=(5,5))
print(demand)
```

```python
# Compute:
# Remaining stock = stock - demand

Remaining_Stock=(stock-demand)
print(Remaining_Stock)
```

```python
# Replace:
# Negative values -> 0(Out of Stock)

Negative_Stock=np.where(Remaining_Stock<0,0,Remaining_Stock)
print(Negative_Stock)
```

```python
# Find:
# Total Shortage

Total_Shortage = np.sum(np.where(Remaining_Stock<0, Remaining_Stock * -1,0))
print(Total_Shortage)
```

```python
# Products with zero stock

Product_shortage=np.sum(np.where(Remaining_Stock == 0,1,0),axis =0)
print(Product_shortage)
```

```python
# Sort each row based on remaining stock

sorted=np.sort(Negative_Stock, axis =1)
print(sorted)
```
