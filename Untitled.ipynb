```python
lst = [1,2,3,'kushagra goel','c',[2,4]]
print(lst)

#append : can help to add another list  from the last.
lst.append([2,3,4])
print(lst)

#extend : can help to add values like if the values are given in list they will be added individually..
lst.extend([2,3,4])
print(lst)

#insert : can help to add values at the given index we wanna add it like if i wanna add something at somehere I have to give it a index then there
# it would be added
lst.insert(2,"divyansh")
print(lst)

# we gonna mostly used insert and append : extend gonna be used very less
```

INSERTING FUNCTIONS

```python
lst = [1,2,3,4,5,'mirai','school']
print(lst)
lst.append([2,3,4])
print(lst)
lst.extend([2,3,4])
print(lst)
lst.insert(3,"divyansh")
print(lst)
```

DELETING FUNCTIONS

```python
lst = [1,2,3,4,'divyansh','mirai']

#pop : helps to remove the last elemnt of the list
lst.pop()
print(lst)

#remove : helps to remove a value if I remove the first value of the value (not the index it removes the first occurance of the given data types)
lst.remove(2)
print(lst)

#index : it removes the index of the value asked
print(lst.index(4))

#count()  this gives the number of values of that element in that list
print(lst.count(4))
```

Sum, min, max, sorted are all functions

```python
lst = [90,87,1001,65,31,1,3]
print(max(lst))
print(min(lst))
a = sorted(lst)
print(a)
print(sum(lst))
```

sort will directly share the list
sorted share the list as a parameter as a whole as it acts as a function in sort the old list would be changed but in case of sorted the old list would not be changed only the new list would be made(the old list would remain unchanged)

TUPLES
We cannot change the value of tuples (immutable) that means the value cannot be added or removed..

```python
tp = (1,3,2,7,4)
print(tp)

a = tuple(sorted(tp))
print(a)
```
